```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter
```

```python
scifi = pd.read_json("results/sci_fi.json")
scifi
```

```python
scifi_count = scifi.count()

print('Sci-Fi attributes with number of records:\n')
print(scifi_count)
```

```python
based_on = []
for column_name, item in scifi.iterrows():
    based_on = based_on + item['based_on']

based_on_count = Counter(based_on)
based_on_count.most_common()
```

```python
genres = []
for column_name, item in scifi.iterrows():
    genres = genres + item['genres']

genres_count = Counter(genres)
genres_count.most_common()
```

```python
origin = []
for column_name, item in scifi.iterrows():
    origin = origin + item['origin']

origin_count = Counter(origin)
origin_count.most_common()
```

```python
year = []
for column_name, item in scifi.iterrows():
    year.append(item['year'])

year_count = Counter(year)
sorted(year_count.most_common())
```

```python
directors = []
for column_name, item in scifi.iterrows():
    directors = directors + item['directors']

directors_count = Counter(directors)
directors_count.most_common()
```

```python
scenarists = []
for column_name, item in scifi.iterrows():
    scenarists = scenarists + item['scenarists']

scenarists_count = Counter(scenarists)
scenarists_count.most_common()
```

```python
actors = []
for column_name, item in scifi.iterrows():
    actors = actors + item['actors']

actors_count = Counter(actors)
actors_count.most_common()
```

```python
musicians = []
for column_name, item in scifi.iterrows():
    musicians = musicians + item['musicians']

musicians_count = Counter(musicians)
musicians_count.most_common()
```
